```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 10)
```

```python
# Fangraphs Roster Resource
fangraphs_df = pd.read_json('https://cdn.fangraphs.com/api/roster-resource/injury-report/data?season=2020')
fangraphs_df.sort_values(by=['date','playerId'], inplace=True)
fangraphs_df.reset_index(drop=True, inplace=True)
fangraphs_df
```

```python
# Spotrac
spotrac_df = pd.read_html('https://www.spotrac.com/mlb/disabled-list/2020/cumulative-player/pitching/')[3]
columns = list()
for col_name in spotrac_df.columns:
    if col_name.startswith('Player'):
        col_name = 'Player'
    elif col_name.startswith('Cash'):
        col_name = 'Earnings'
    columns.append(col_name)
spotrac_df.columns = columns
spotrac_df
```

```python
spotrac_df.dropna(subset=['Player'], inplace=True)
spotrac_df['Name'] = spotrac_df['Player'].str.extract(r'(.*)\s,\s.*')
spotrac_df.drop(['Player', 'Pos.', 'Status'], axis='columns', inplace=True)
spotrac_df
```

```python
spotrac_df['Injury1'] = spotrac_df['Injury'].str.split('/').str.get(0)
spotrac_df['Injury2'] = spotrac_df['Injury'].str.split('/').str.get(1)
spotrac_df['Injury3'] = spotrac_df['Injury'].str.split('/').str.get(2)
spotrac_df['StartDate1'] = spotrac_df['Dates'].str.extract(r'(\d{1,2}/\d{1,2})\s-\s.*')
spotrac_df['EndDate1'] = spotrac_df['Dates'].str.extract(r'\d{1,2}/\d{1,2}\s-\s(\d{1,2}/\d{1,2}).*')
spotrac_df['StartDate2'] = spotrac_df['Dates'].str.extract(r'\d{1,2}/\d{1,2}\s-\s\d{1,2}/\d{1,2}(\d{1,2}/\d{1,2})\s-\s.*')
spotrac_df['EndDate2'] = spotrac_df['Dates'].str.extract(r'\d{1,2}/\d{1,2}\s-\s\d{1,2}/\d{1,2}\d{1,2}/\d{1,2}\s-\s(\d{1,2}/\d{1,2}).*')
spotrac_df['StartDate3'] = spotrac_df['Dates'].str.extract(r'\d{1,2}/\d{1,2}\s-\s\d{1,2}/\d{1,2}\d{1,2}/\d{1,2}\s-\s\d{1,2}/\d{1,2}(\d{1,2}/\d{1,2})\s-\s.*')
spotrac_df['EndDate3'] = spotrac_df['Dates'].str.extract(r'\d{1,2}/\d{1,2}\s-\s\d{1,2}/\d{1,2}\d{1,2}/\d{1,2}\s-\s\d{1,2}/\d{1,2}\d{1,2}/\d{1,2}\s-\s(\d{1,2}/\d{1,2}).*')
spotrac_df.sort_values(['StartDate3','StartDate2','StartDate1'])
```
